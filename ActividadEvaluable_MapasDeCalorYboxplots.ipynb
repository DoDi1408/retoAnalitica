```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
df = pd.read_csv("playlist_2010to2022.csv")
```

```python
fig = plt.figure(figsize=(11,11))
corr = df.corr(numeric_only=True)
sns.heatmap(data=corr,vmin = -1, vmax=1,cmap = "YlOrRd", annot=True)
plt.title("Heatmap Spotify")
fig.show()
```

```python
fig = plt.figure(figsize=(11,11))
sns.boxplot(data = df, y = "loudness", x = "year")
plt.title("BoxPlot Loudness by Year")
fig.show()
```

```python
fig = plt.figure(figsize=(11,11))
sns.boxplot(data = df, y = "energy", x = "year")
plt.title("BoxPlot energy by Year")
fig.show()
```

```python
fig = plt.figure(figsize=(11,11))
sns.boxplot(data = df, y = "danceability", x = "year")
plt.title("BoxPlot Danceability by Year")
fig.show()
```

```python
fig = plt.figure(figsize=(11,11))
#df["duraton_ms"] = df["duration_ms"]/60000
sns.boxplot(data = df, y = "duration_ms", x = "year")
plt.title("BoxPlot Duration by Year")
fig.show()
```

```python
fig = plt.figure(figsize=(11,11))
sns.histplot(data=df, x="year", y="artist_popularity",shrink=.8)
plt.title("Histogram Artist Popularity of Top Songs by Year")
fig.show()
```
